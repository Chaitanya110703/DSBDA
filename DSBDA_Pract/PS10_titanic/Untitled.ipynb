```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

## Scatter Plot

```python
sns.scatterplot(data = df, x = 'Age', y= 'Fare')
plt.show()
```

```python
sns.scatterplot(data = df, x = 'Age', y= 'Fare', hue = 'Sex')
plt.show()
```

## HistPlot

```python
sns.histplot(data = df,x ='Age')
plt.show()
```

```python
sns.histplot(data = df,x ='Age',hue = 'Pclass')
plt.show()
```

```python
sns.displot(data = df,x ='Age',hue = 'Sex',bins = 30)
plt.show()
```

```python
sns.displot(data = df,x ='Age',hue = 'Sex',bins = 30, col='Embarked')
plt.show()
```

```python
sns.barplot(data = df, x = 'Sex', y= 'Age')
plt.show()
```

## CountPlot

```python
sns.countplot(df, x = 'Pclass',hue = 'Sex')
plt.show()
```

```python
sns.countplot(df, x = 'Pclass',hue = 'Embarked')
plt.show()
```

```python
sns.countplot(df, x = 'Pclass',hue = 'Survived')
plt.show()
```

## Catplot

```python
sns.catplot(x='Embarked',hue='Survived',data=df,kind='count')
plt.show()
```

```python
sns.catplot(x='Embarked',hue='Survived',data=df,kind='count', col = 'Pclass')
plt.show()
```
