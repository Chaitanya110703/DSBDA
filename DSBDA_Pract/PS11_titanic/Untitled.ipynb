### 1. Use the dataset 'titanic'. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names : 'sex' and 'age')
### 2. Write observations on the inference from the above statistics. Use violin plot and others plots from seaborn package

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df
```

## Boxplot

```python
sns.boxplot(df, x = 'Sex', y = 'Age')
```

#### Box plots visually show the distribution of numerical data and skewness through displaying the data quartiles (or percentiles) and averages.


```python
sns.boxplot(df, x = 'Age', y = 'Sex')
```

```python
sns.boxplot(df, x = 'Sex', y = 'Age', hue = 'Survived')
plt.show()
```

## Violin Plot

```python
sns.violinplot(df , x = 'Sex', y = 'Age')
```

```python
sns.violinplot(df , x = 'Sex', y = 'Age', hue = 'Survived')
```

```python
sns.violinplot(df , x = 'Sex', y = 'Age', hue = 'Survived', split=True)
```

## StripPlot

```python
sns.stripplot(x = 'Sex', y = 'Age',data = df)
```

```python
sns.stripplot(x = 'Sex', y = 'Age',data = df, hue='Survived',palette='Set1', dodge=True)
plt.show()
```

## Swarm Plot

```python
sns.swarmplot(x = 'Age', y = 'Sex', data = df, palette='Set1')
plt.show()
```

```python
sns.swarmplot(x = 'Sex', y = 'Age', data = df, palette='Set2')
plt.show()
```

```python
sns.swarmplot(x = 'Age', y = 'Sex', data = df, palette='Set1', hue = 'Survived')
plt.show()
```

```python
sns.swarmplot(x = 'Sex', y = 'Age', data = df, palette='Set1', hue='Survived')
plt.show()
```

```python
sns.swarmplot(x = 'Sex', y = 'Age', data = df, palette='Set1', hue='Survived', dodge=True, size = 3)
plt.show()
```
